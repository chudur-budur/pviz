```python
%reload_ext autoreload
%autoreload 2

import sys
import os

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams.update({'figure.max_open_warning': 0})
```

```python
randfpath = "../data/gaa/10d/dataf.csv"
nbifpath = "../data/gaa-nbi/10d/dataf.csv"


F = np.loadtxt(nbifpath, delimiter=',')
Fr = np.loadtxt(randfpath, delimiter=',')

print(F.shape)
print(Fr.shape)

Id = np.arange(0,F.shape[0],1).astype(int)
I0 = np.zeros(F.shape[0]).astype(int)
I1 = np.ones(F.shape[0]).astype(int) 


If = np.column_stack((Id,I0))
Ir = np.column_stack((Id,I1))
Ia = np.row_stack((If,Ir))
np.random.shuffle(Ia)
print(Ia.shape)

I0m = np.where(Ia[:,1] == 0)
I1m = np.where(Ia[:,1] == 1)
print(I0m)
print(I1m)
print(np.intersect1d(I0m, I1m))

Fm = np.zeros((Ia.shape[0],F.shape[1]))
Fm[I0m,:] = F
Fm[I1m,:] = Fr
print(Fm)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(Fm[:,0], Fm[:,1], Fm[:,2], s=1)
plt.show()
```
